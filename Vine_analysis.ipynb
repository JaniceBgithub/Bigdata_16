```python
import pandas as pd
import csv
```

```python
file_path = "https://kitchen12.s3.us-east-2.amazonaws.com/vine_data2.csv"

#df = pd.read_csv("vine_data2.csv")
df = pd.read_csv(file_path)
df.head(10)
```

```python
df = df[df["total_votes"]>=20]
df.head(20)
```

```python
df['helpful_percent'] = df["helpful_votes"] / df["total_votes"] *100
df.head(50)

```

```python
df2 = df[df["helpful_percent"]>=50]
df2.head(25)
```

```python
df_paid = df2[df2["vine"]=='Y']
df_unpaid = df2[df2["vine"]=='N']
```

```python
paid_reviews =  len(df_paid.index)
paid_reviews
```

```python
unpaid_reviews = len(df_unpaid.index)
unpaid_reviews
```

```python
paid_5star = df_paid[(df_paid['star_rating']==5)].count()
unpaid_5star = df_unpaid[(df_unpaid['star_rating']==5)].count()
```

```python
paid_5star
```

```python
unpaid_5star
```

```python
percent_5star_paid = paid_5star / paid_reviews * 100
percent_5star_unpaid = unpaid_5star / unpaid_reviews * 100
```
